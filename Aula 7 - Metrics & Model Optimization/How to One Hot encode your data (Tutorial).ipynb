## Example of Binary Encoding
- Manual Mapping for 2 categories

```python
df['Gender'].value_counts()
```

We have 2 different classes, so we can simply replace the FEMALE by 1 and the MALE by 0, or the opposite!

```python
df['Gender'] = df['Gender'].replace({'FEMALE':1, 'MALE':0})
```

```python
df.head(3)
```

```python
df["Gender"].value_counts()
```

## Example of One Hot Encoding using get_dummies function

```python
df["Tumour_Stage"].value_counts()
```

For the column "Tumour_stage" we have 3 different categories, so we should apply onehotencoding using the get_dummies function from the pandas .
This function will remove the original columns and create new columns in the end of the dataframe. We set drop_first=True, so it automatically removes the first column since with 3 categories, only 2 are enough to explain the 3 classes

```python
df_encoded = pd.get_dummies(df, columns=['Tumour_Stage'], drop_first=True, dtype=int)
df_encoded.head()
```

```python
df_encoded["Tumour_Stage_II"].value_counts()
```

```python
df_encoded["Tumour_Stage_III"].value_counts()
```
