# **Data Cleaning**

Missing values can be handled using the following methods, listed by priority:

1.   Drop rows where all values are NaN
     *   df.dropna(how='all')
     *   Not applicable here, as there are no completely empty rows in the dataset

2.   Drop rows with missing values in key columns
     *   df.dropna(subset=['column'])
     *   Key columns have not been identified yet - thus, maintain integrity of all independent variables

3.   Drop rows with any missing values
     *   df.dropna()
     *   Simple and reasonable choice since it does not introduce significant bias (proportion of CHD cases remains consistent)

4.   Fill with the mean / median / mode
     *   df['column'].fillna(df['column']).mean()
     *   df['column'].fillna(df['column']).median()
     *   df['column'].fillna(df['column']).mode()
     *   Preserves dataset size but may introduce artificial patterns and increase the risk of overfitting

5.   Fill with a constant
     *   df.fillna()
     *   Risk of introducing misleading or biased information

#Pearson Correlation

#### Why use Pearson correlation to detect 10-year CHD risk?

Pearson correlation helps us identify **which clinical and lifestyle variables show linear associations with the development of Coronary Heart Disease (CHD)** over time.

- Variables such as **age, blood pressure, cholesterol, and body mass index** often have a linear relationship with the risk of developing CHD.
- By calculating the correlation between these variables and the target variable (`TenYearCHD` — indicating whether the patient developed CHD within 10 years), we gain an initial insight into the most relevant features.

This is useful for:

- **Selecting important variables** for predictive models, such as logistic regression or machine learning techniques.
- **Understanding the direction and strength of associations** (positive or negative).
- **Reducing dimensionality** by removing weakly correlated variables.

> **Important:** Pearson correlation only measures linear relationships and **does not imply causation**. Therefore, it should be used as an initial exploratory analysis, followed by more sophisticated models for reliable prediction.


```python
import pandas as pd  # Para trabalhar com tabelas de dados (DataFrames)
import numpy as np   # Para fazer operações matemáticas, especialmente com números
import seaborn as sns  # Para deixar os gráficos mais bonitos e fáceis de entender
import matplotlib.pyplot as plt  # Para desenhar gráficos no computador
from IPython.display import display, Markdown  # Para mostrar textos formatados bonitos no notebook

# Configura o estilo dos gráficos para ficarem com fundo quadriculado (tipo caderno) e define o tamanho padrão dos gráficos
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 5)

# Função que mostra quais variáveis têm correlações fortes (parecem muito entre si)
def mostrar_correlacoes_fortes(corr_matrix, variaveis_numericas, limiar=0.5):
    # Começa uma lista de linhas que serão usadas para criar uma tabela bonitinha
    linhas = []
    linhas.append("### Strong Correlations (|r| > 0.5)\n")  # Título da tabela
    linhas.append("| Variable 1       | Variable 2       | Correlation (r) |")  # Cabeçalho da tabela
    linhas.append("|------------------|------------------|----------------|")  # Linha separadora do cabeçalho

    # Percorre todas as combinações possíveis de variáveis numéricas para verificar correlações fortes
    for i in range(len(variaveis_numericas)):
        for j in range(i+1, len(variaveis_numericas)):
            r = corr_matrix.iloc[i, j]  # Pega o valor da correlação entre as variáveis i e j
            if abs(r) > limiar:  # Se a correlação for maior que o limite definido (padrão 0.5)
                # Adiciona uma linha na tabela mostrando as duas variáveis e o valor da correlação
                linhas.append(f"| `{variaveis_numericas[i]:<15}` | `{variaveis_numericas[j]:<15}` | {r:>14.2f} |")

    linhas.append("\n---\n")  # Linha de separação após a tabela

    # Texto explicando o que significa cada variável, para ajudar a entender a tabela
    legend_text = """
### Caption

| Column            | Description                                                              |
|-------------------|---------------------------------------------------------------------------|
| `male`            | Patient's gender (1 = male, 0 = female)                                   |
| `age`             | Patient's age in years                                                    |
| `education`       | Level of education (higher values indicate higher education level)        |
| `currentSmoker`   | Whether the patient is a smoker (1 = yes, 0 = no)                         |
| `cigsPerDay`      | Average number of cigarettes smoked per day                               |
| `BPMeds`          | Use of blood pressure medication (1 = yes, 0 = no)                        |
| `prevalentStroke` | History of stroke (1 = yes, 0 = no)                                       |
| `prevalentHyp`    | History of hypertension (1 = yes, 0 = no)                                 |
| `diabetes`        | Presence of diabetes (1 = yes, 0 = no)                                    |
| `totChol`         | Total blood cholesterol (mg/dL)                                           |
| `sysBP`           | Systolic blood pressure (mmHg)                                            |
| `diaBP`           | Diastolic blood pressure (mmHg)                                           |
| `BMI`             | Body Mass Index (kg/m²)                                                   |
| `heartRate`       | Heart rate (beats per minute)                                             |
| `glucose`         | Blood glucose level (mg/dL)                                               |
"""
    linhas.append(legend_text)  # Adiciona essa explicação no final

    # Mostra todo o texto formatado como Markdown bonitinho no notebook
    display(Markdown("\n".join(linhas)))


# Função que cria e mostra a matriz de correlação de Pearson, com gráfico e resumo das correlações fortes
def criar_matrix_correlacao_pearson(df, title='Pearson Correlation Matrix', drop_cols=None, show_summary=True):
    """
    Plota uma matriz de correlação de Pearson e exibe um resumo das correlações fortes.

    Parâmetros:
    - df: DataFrame com os dados
    - title: Título do gráfico
    - drop_cols: Lista de colunas para remover antes do cálculo (ex: variável alvo)
    - show_summary: Se True, mostra o resumo das correlações fortes e a legenda das variáveis
    """

    # Se for passado alguma coluna para excluir, tira ela do DataFrame (se existir)
    if drop_cols:
        df = df.drop(columns=drop_cols, errors='ignore')

    # Seleciona apenas as colunas que são números para fazer a correlação
    df_numeric = df.select_dtypes(include=[np.number])
    # Calcula a matriz de correlação usando o método Pearson
    corr = df_numeric.corr(method='pearson')

    # Cria um gráfico para mostrar essa matriz de correlação
    fig, ax = plt.subplots(figsize=(12, 10))
    # Desenha a matriz como uma imagem colorida, com cores mostrando a força e direção da correlação
    cax = ax.matshow(corr, cmap='coolwarm')
    # Coloca o título no gráfico
    plt.title(title, fontsize=16, pad=20)

    # Define os nomes das colunas na horizontal e vertical do gráfico
    ax.set_xticks(np.arange(len(corr.columns)))
    ax.set_yticks(np.arange(len(corr.columns)))
    ax.set_xticklabels(corr.columns, rotation=90)
    ax.set_yticklabels(corr.columns)

    # Coloca os valores numéricos da correlação em cada quadradinho do gráfico
    for i in range(len(corr.columns)):
        for j in range(len(corr.columns)):
            ax.text(j, i, f'{corr.iloc[i, j]:.2f}',
                    ha='center', va='center', color='black', fontsize=8)

    # Adiciona uma barra lateral de cores para ajudar a entender os valores
    fig.colorbar(cax)
    plt.tight_layout()  # Ajusta para o gráfico não ficar cortado
    plt.show()  # Mostra o gráfico

    # Se o parâmetro show_summary for True, mostra as correlações fortes e legenda usando a função criada antes
    if show_summary:
        variaveis_numericas = df_numeric.columns.tolist()
        mostrar_correlacoes_fortes(corr, variaveis_numericas)


# Carrega a tabela de dados de um link na internet e coloca dentro do computador
url = 'https://raw.githubusercontent.com/Hospital-Da-Luz-Learning-Health/MLCatolica25/refs/heads/main/Final%20Project/data/train.csv'
df = pd.read_csv(url)

# Chama a função para criar o gráfico da matriz de correlação, tirando a variável 'TenYearCHD' para não atrapalhar
criar_matrix_correlacao_pearson(df, drop_cols=['TenYearCHD'])
```

#### Pearson Correlation Interpretation
Note the correlation between **diastolic** and **systolic** blood pressure is clinically significant, and both measures individually play a role in predicting the risk of coronary heart disease (CHD) within 10 years. Systolic BP is more strongly associated with CHD risk than Diastolic BP, especially in adults over 50. Moreover, Systolic blood pressure (SBP) is a better predictor of cardiovascular events than diastolic, particularly in older adults. Elevated SBP increases arterial wall stress, promoting atherosclerosis and end-organ damage.
On the other hand,Diastolic BP still matters — especially in younger adults. In younger populations, elevated diastolic BP (≥90 mmHg) is predictive of future cardiovascular risk. Very low diastolic BP (e.g., <60 mmHg), particularly with high SBP (widened pulse pressure), can also be a marker of arterial stiffness and is associated with poor outcomes.

Both **smoking** and **age** are independent risk factors for CHD, as Smoking and Age increases CHD risk. TheA negative correlation between them means the younger people in your sample have higher smoking rates, which could amplify their CHD risk for their age group. Older adults may have lower smoking rates, which slightly moderates their CHD risk, but their age alone still keeps risk elevated.

# Chi-square

#### Why use Chi-square tests to detect 10-year CHD risk?
Chi-square tests are used to examine the association between two categorical variables. In this dataset, several variables such as gender (male), smoking status (currentSmoker), diabetes (diabetes), and medication use (BPMeds) are binary or categorical. The Chi-square test allows me to determine whether there is a statistically significant relationship between these categorical variables, such as whether the prevalence of hypertension is associated with smoking status or diabetes. This helps identify potential interactions and informs feature selection for later predictive modeling of cardiovascular disease risk.

```python
import pandas as pd  # Importa uma biblioteca que ajuda a organizar e analisar dados em tabelas.
from scipy.stats import chi2_contingency  # Importa uma função para fazer um teste estatístico chamado qui-quadrado, que ajuda a ver se duas coisas estão relacionadas.
from IPython.display import display  # Importa uma função para mostrar tabelas de um jeito mais bonito na tela.

# Função que cria um gráfico colorido para mostrar os resultados do teste qui-quadrado
def plot_chi_square_pvalues(chi_results, variables):
    # Cria uma tabela vazia para guardar os valores de p (p-values) para cada par de variáveis
    matrix = pd.DataFrame(index=variables, columns=variables)

    # Para cada resultado do teste, pega os nomes das variáveis e o p-valor, e coloca na tabela nas posições certas
    for _, row in chi_results.iterrows():
        v1 = row['Variable 1']
        v2 = row['Variable 2']
        pval = row['p-value']
        matrix.loc[v1, v2] = pval  # Coloca o p-valor na posição (linha v1, coluna v2)
        matrix.loc[v2, v1] = pval  # Também coloca na posição invertida para deixar a tabela simétrica

    # Na diagonal da tabela (quando a variável é comparada consigo mesma), colocamos vazio porque não faz sentido comparar ela com ela mesma
    for v in variables:
        matrix.loc[v, v] = None

    matrix = matrix.astype(float)  # Converte os valores para números decimais para garantir que o gráfico funcione direito

    # Aqui desenha o gráfico colorido que mostra os p-valores, com uma legenda e título
    plt.figure(figsize=(8, 6))
    sns.heatmap(matrix, annot=True, fmt=".4f", cmap="RdYlGn_r",
                cbar_kws={'label': 'p-value'}, linewidths=0.5, linecolor='gray')
    plt.title("Heatmap dos p-values (Teste Qui-Quadrado)")
    plt.tight_layout()
    plt.show()

# Função que faz o teste qui-quadrado para todos os pares possíveis de variáveis na lista
def run_chi_square_tests(df, variables):
    results = []  # Começa uma lista para guardar os resultados do teste

    # Vai passar por cada variável na lista (de i até o fim)
    for i in range(len(variables)):
        # Para cada variável, compara com as outras que ainda não foram comparadas (para não repetir)
        for j in range(i + 1, len(variables)):
            var1 = variables[i]  # Pega a variável na posição i
            var2 = variables[j]  # Pega a variável na posição j

            # Cria uma tabela que mostra quantas vezes cada combinação de valores das duas variáveis acontece
            table = pd.crosstab(df[var1], df[var2])

            # Aplica o teste qui-quadrado nessa tabela para saber se as duas variáveis têm alguma relação
            chi2, p, dof, expected = chi2_contingency(table)

            # Guarda o resultado: nomes das variáveis, valor do teste qui-quadrado, p-valor e se é estatisticamente significativo (p < 0.05)
            results.append({
                'Variable 1': var1,
                'Variable 2': var2,
                'Chi2 Stat': chi2,
                'p-value': p,
                'Significant (p < 0.05)': p < 0.05
            })

    # Transforma a lista de resultados em uma tabela para facilitar a visualização
    return pd.DataFrame(results)

# Função para mostrar os resultados do teste de um jeito visual, com cores que indicam se o resultado é importante ou não
def display_chi_square_results(results_df):
    # Pinta de verde os p-valores menores que 0.05 (que significam relação importante), e de vermelho os maiores
    def highlight_p_value(val):
        return 'background-color: #d4edda' if val < 0.05 else 'background-color: #f8d7da'

    # Pinta de verde as linhas que são significativas, e vermelho as que não são
    def highlight_significance(val):
        return 'background-color: #d4edda' if val else 'background-color: #f8d7da'

    # Aplica essas cores, formata os números para mostrar com algumas casas decimais, e prepara a tabela para mostrar
    styled = results_df.style \
        .background_gradient(subset=["Chi2 Stat"], cmap="Blues") \
        .format({
            "Chi2 Stat": "{:.3f}",
            "p-value": "{:.4f}"
        }) \
        .applymap(highlight_p_value, subset=["p-value"]) \
        .applymap(highlight_significance, subset=["Significant (p < 0.05)"])

    display(styled)  # Mostra a tabela colorida na tela

# Baixa um arquivo com dados para usar
url = 'https://raw.githubusercontent.com/Hospital-Da-Luz-Learning-Health/MLCatolica25/refs/heads/main/Final%20Project/data/train.csv'
df = pd.read_csv(url)  # Lê o arquivo e cria uma tabela com os dados

# Lista as variáveis que são do tipo "sim ou não" (categorias binárias) para testar
cat_vars = ['male', 'currentSmoker', 'BPMeds', 'prevalentStroke', 'prevalentHyp', 'diabetes']

# Roda o teste qui-quadrado para todos os pares dessas variáveis
chi_results = run_chi_square_tests(df, cat_vars)

# Mostra os resultados com cores para facilitar a interpretação
display_chi_square_results(chi_results)

# Desenha o gráfico colorido mostrando os p-valores de cada par de variáveis
plot_chi_square_pvalues(chi_results, cat_vars)
```
