# Selecting data (loc and iloc)

## .loc

`loc` is short for locate

```python
df.loc[:, :] # isto seleciona todas as linhas e colunas (redundante, sim)
```

The `...` strike again in the output above. Again, it's just a display optimization, the 303 rows are still there.

```python
df.loc[:, 'age'] # devolve todos os valores da coluna "age", é o mesmo que fazer df['age']
```

```python
df.loc[:,['age','slope']] # devolve todos os valores das colunas "age" e "slope", é o mesmo que fazer df[['age','slope']]
```

```python
df.loc[23,:] # devolve toda a informação do paciente 23
```

```python
df.loc[23, 'age'] # devolve a idade do paciente 23
```

We can create programs around this data:

```python
if df.loc[23, 'age'] > 50:
    print("O paciente com id 23 tem mais de 50 anos.")
```

We can select multiple patients

```python
my_patients = [23, 185]

df.loc[my_patients, :] # devolve toda a informação dos pacientes 23 e 185
```

The above is the same as doing:

```python
df.loc[[23, 185], :]
```

And to select multiple patients and multiple columns:

```python
df.loc[[44, 21], ['restecg', "ca"]]
```

## .iloc

With iloc you select by the index position of the rows and/or column and not their values.

```python
df.iloc[0,:] # devolve toda a informação do doente na primeira posição da dataframe
```

```python
df.iloc[:,0] # devolve a primeira coluna do DataFrame de todos os doentes
```

```python
df.iloc[[0,10], [0,1]] # devolve a primeira e segunda coluna dos doentes na primeira e décima posição do DataFrame
```
